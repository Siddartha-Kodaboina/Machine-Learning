### 1. Distance between 2 vectors



*   A vector of length-n represents a point in n-dimentional space
*   $\text({x_1}, x_2\text)$ is in 2-D, $\text({x_1}, x_1\text, x_3 )$ is in 3-D and  $\text({x_1}, x_1\text, x_3, ......x_n )$ is in n-D.

The distance from origin to a point P is

distance d = $\sqrt{∑_{i=1}^n x_i^2}$

Where, n is the dimentions of the space.



### 2. Product of two vectors

$A.B = \sum_{i=1}^n a_i b_i = |a| |b| cosθ$

where, 

A = [a1, a2, a3, ..., an]

B = [b1, b2, b3, ..., bn]

|a| = distance of point A from origin =  $\sqrt{∑_{i=1}^n a_i^2}$

|b| = distance of point B from origin =  $\sqrt{∑_{i=1}^n b_i^2}$
