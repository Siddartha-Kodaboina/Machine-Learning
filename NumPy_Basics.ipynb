Different ways of creating numpy arrays ( 1-D, 2-D, 3-D, ... , n-D)

1. np.arange() => https://numpy.org/doc/stable/reference/generated/numpy.arange.html

```python
# 1. Explicitely providing the values
a = np.array([1, 2, 3, 4])
print(a)
```

```python

# 2. Using arange() function
# arange(n) => creates a vector of len n, values are from 0 -> (n-1)
arrange_array = np.arange(10)
print(arrange_array)
```

```python
# 3. Using np.arange() with a range=> arange(x, y)
arrange_array_range = np.arange(2, 10)
print(arrange_array_range)
```

```python
# 4. Using np.arange() with a range and step=> arange(x, y, step)
arrange_array_range_step = np.arange(2, 10, 3)
print(arrange_array_range_step)
```

```python
# 5. using np.linspace() 
# linspace(x, y, number_of_points) => same as of arrange(), except it doesn't have step concept, instead it provides a vector of length number_of_points in which ar[i+1] - ar[i] is same for all elements
print("\nUsing np.linspace() \n")
linspace_array = np.linspace(2, 10, 5, dtype=int, axis=-1, retstep=True )
print(linspace_array)
```

```python
# 6. Using np.ones() 1-d & 2-d & M-D
print("\nUsing np.ones() \n")
ones_array = np.ones((3), dtype=int) # 1-D
print(ones_array)
ones_array_2 = np.ones((3, 2), dtype=int) # 2-D
print(ones_array_2)
ones_array_3 = np.ones((4, 3, 2), dtype=int) # 3-D
print(ones_array_3)

```

```python
# 7. Using np.zeros() 1-d & 2-d & M-D
print("\nUsing np.zeros() \n")
zeros_array = np.zeros((3), dtype=int) # 1-D
print(zeros_array)
zeros_array_2 = np.zeros((3, 2), dtype=int) # 2-D
print(zeros_array_2)
zeros_array_3 = np.zeros((4, 3, 2), dtype=int) # 3-D
print(zeros_array_3)

```

```python
# 8. Using np.eye() 1-d & 2-d & M-D
# It creates a identical matrix( it is a matrix in which all major diagonal elements are 1's and rest are 0's )
print("\nUsing np.eye() \n")
eye_array = np.eye(3, dtype=int) # 3 * 3 matrix
print(eye_array)
eye_array_2 = np.eye(3, 2, dtype=int) # 3 * 2 matrix
print(eye_array_2)
eye_array_3 = np.eye(4, 3, -1, dtype=int) # 4 * 3 matrix
print(eye_array_3)
```

```python
# 9. Creating a diagonal array
diag_arange_array = np.ones((1,4))
print(diag_arange_array)
diag_array = np.diag(diag_arange_array)
print(diag_array)
```

Slicing 
1. It makes the original array(a) and subarray(b) to share the memory for the memory optimization

```python
a = np.arange(10)
print(a)

# slicing
print(a[1:7])

# slicing with step 
b = a[1:7:2]
print(b)

# reversing array with slicing
b = a[::-1]
print(b)
```

# Copies and views

```python
np.shares_memory(a, b) # True if a and b shares memory , otherwise it is false

b = a[0::2]
# Changing a value at array b, changes the  value in array a
print(a)
print(b)

b[0] = 1
print(a)
print(b)
```

**Force copy**

As, slicing makes the subarray to share the memory of original array. So, We cannot change the value in subarray, because changing a value in the subarray changes the value in original array. For that this .copy() is useful. It creates new independent memory for sliced subarray.

```python
print(np.shares_memory(a, b)) # True if a and b shares memory , otherwise it is false

b = a[0::2].copy()
# Changing a value at array b, doesn't changes the  value in array a
print(a)
print(b)

print(np.shares_memory(a, b))
b[0] = 1
print(a)
print(b)
```

**Indexing and Mask**

```python
a = np.random.randint(0, 20, 15)
print(a)
```

```python

# mask is a filter, we apply to a ndarray to filter out and to create a sub-array
mask = (a%3 == 0)
print(mask)

# Mask creates force copy, doesn't create a view
# which means, mask will have their own memory and independent of original array
filtered_array = a[mask]
print(filtered_array)

a[mask] = -1
print(a)



```
