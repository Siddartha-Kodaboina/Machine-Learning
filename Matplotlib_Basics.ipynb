```python
 # importing libraries
 import matplotlib.pyplot as plt
 import numpy as np
```

```python
x = np.arange(-5000, 5000)
# y = x**2
# print(x)
# print(y)
```

```python
plt.plot(x, x**2)
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.title("Numbers vs Their Squares")
plt.grid()  # for adding the grid
plt.show()
```

```python
plt.plot(x, x**3)
plt.xlabel("Numbers")
plt.ylabel("Cubes")
plt.title("Numbers vs Their Cubes")
plt.grid()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
plt.suptitle("Numbers vs Their Exponents")

ax1.plot(x, x**2)
ax2.plot(x, x**3)
```

Multiple plots on a single plot object

Labelling & Adding symbols and colors to each plot

```python
x = np.arange(1, 10, 0.2)

plt.plot(x, x**2, 'co', label="^2")
plt.plot(x, x**2.3, 'r^', label="^2.3")
plt.plot(x, x**2.6, 'bs', label="^2.6")

plt.grid(True)
plt.legend()
plt.show()
```
