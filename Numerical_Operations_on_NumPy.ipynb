### Comaparisons

```python
# element-wise comparisons ( to compare every element of the array with every element of the other array i.e., A[i] compared with B[i], A[i+1] compared with B[i+1], .. A[i+n] compared with B[i+n] ... and so on)

a = np.random.randint(1, 20, 6)
b = np.random.randint(1, 20, 6)
c = np.random.randint(1, 20, 6)
print(a)
print(b)
print(c)

# We can do all logical operations here, <, >, <=, >=, != all...
print(a == b)
print(b == c)
print(a == c)

```

```python
# Array-wise comparisons (All elements of an array compared with all the elements of other array, if all elements matches the condition the True else False)

d = a.copy()
# d= np.random.randint(1, 20, 5)
# np.shares_memory(a, d)

np.array_equal(a, d)
```

### Mathematical Functions

```python
a = np.random.randint(1, 20, 7)
a
```

```python
print(np.sin(a))
print(np.log(a))
print(np.exp(a))
```

### Array Reductions

```python
a = np.arange(1, 20, np.sum(np.random.randint(6, 10, 1)))
a
```

```python
np.sum(a) # Sum of all the elements in numpy array a

maxi = a.max() # returns the max element of the numpy array a
print(maxi)

maxi_ind = a.argmax() # returns the index of maximum element in numpy array a
print(maxi_ind)

mini = a.min() # returns the min element of the numpy array a
print(mini)

mini_ind = a.argmin() # returns the index of min element in numpy array a
print(mini_ind)

```
